### Q1

```python
import matplotlib.pyplot as plt
import numpy as np
import time
```

#### a

```python

def f(vec):
    return (np.sin(np.sqrt(100)*vec))**2

vec = np.linspace(0,1,1001)
plt.plot(vec,f(vec))
plt.show()
```

#### b

```python
def I_func(dx):
    I = 0
    vec = np.arange(0,1,dx)
    N = len(vec)-1
    
    for n in range(N):
        I += dx/2.0 * (f(vec[n]) + f(vec[n+1]))
    return I
    
def eps(Ii,Iii):
    return np.abs((1/3.0 * (Ii-Iii)))

start_time = time.perf_counter()

dx = vec[-1] - vec[0]
I = 0
eps_val = np.inf
Iii = I_func(dx)
while eps_val>1e-6:
    dx = dx/2 
    Ii = I_func(dx)
    eps_val = eps(Ii,Iii)
    Iii = Ii
    
end_time = time.perf_counter()

# Calculate the elapsed time
elapsed_time = end_time - start_time
print(f"epsilon {eps_val:.3e} ")
print(f"The operation completed in {elapsed_time:.4f} seconds")
print(f"Area is :{Iii:.6f}")
print(f"dx is :{dx:.4e}")
```

```python
def I1_f(dx):
    return dx/2.0 * (f(0) + f(dx))

def Ii_f(dx,I_prev):
    N = len(np.arange(0,1,dx)) - 1
    I  = 0.5 * I_prev
    for i in np.arange(0,N,2):
        I += dx*np.sum(f(0+i*dx))
    return I

def eps_fun(I,I_prev):
    return np.abs((1/3.0 * (I-I_prev)))

dx = 1
I1 = I1_f(1)
dx = dx/2

I2 = Ii_f(dx,I1)
I_prev = I2
eps_val = eps_fun(I2,I1)
start_time = time.perf_counter()
while eps_val>1e-6:
    dx = dx/2 
    I  = Ii_f(dx,I_prev)
    eps_val = eps_fun(I,I_prev)
    I_prev = I


end_time = time.perf_counter()

# Calculate the elapsed time
elapsed_time = end_time - start_time
print(f"epsilon {eps_val:.3e} ")
print(f"The operation completed in {elapsed_time:.4f} seconds")
print(f"Area is :{Iii:.6f}")
print(f"dx is :{dx:.4e}")
```

#### c

```python
def R11_f(dx):
    return dx/2.0 * (f(0) + f(dx))

def Ri1_f(dx,Rim1i):
    N = len(np.arange(0,1,dx)) - 1
    I  = 0.5 * Rim1i
    for i in np.arange(0,N,2):
        I += dx*np.sum(f(0+i*dx))
    return I


def R_imp1_f (Rim,Rim1m,m):
    return Rim + 1/(4**m - 1) * (Rim-Rim1m)

def eps_fun (Rii,Rim1m,I):
    return np.abs(1/(4**I-1)*(Rii-Rim1m))


I=1
dx = 1
R11 = R11_f(dx)
dx = dx/2

I+=1
R21 = Ri1_f(dx,R11)
R22 = R_imp1_f(R21,R11,3)
eps_val = eps_fun(R22,R11,I)

R_prev = [R21,R22]

R = list()
start_time = time.perf_counter()
while eps_val>1e-6:
    I+=1
    dx=dx/2
    R = [Ri1_f(dx,R_prev[-1])]
    for m in range(I-1):
        R.append(R_imp1_f(R[-1],R_prev[m],m+1))
    eps_val = eps_fun(R[-1],R_prev[-1],I)
    R_prev = R
    


end_time = time.perf_counter()

# Calculate the elapsed time
elapsed_time = end_time - start_time
print(f"epsilon {eps_val:.3e} ")
print(f"The operation completed in {elapsed_time:.4f} seconds")
print(f"Area is :{R[-1]:.6f}")
print(f"dx is :{dx:.4e}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
def func(x):
    return np.sin((x-x**2)/x)
    
def noisy_func(x,lamb):
    return func(x) + lamb*np.random.uniform(low=-1.0, high=1.0, size=x.shape)
vec = np.linspace(0.5,10,1001)
f = func(vec)  
noisy_f = noisy_func(vec,1)
plt.plot(vec,noisy_f)
plt.plot(vec,f)
plt.show()
```

#### b

```python
vec = np.linspace(0.5,10,1001)
dx = vec[2:-1] - vec[0:-3]
f = func(vec)  
lambda_vec = np.array([2,1,0.5,0.1])
f_tag = (f[0:-3]-f[2:-1])/dx
plt.plot(vec[1:-2],f_tag)
for l in lambda_vec:
    noisy_f = noisy_func(vec,l)
    noisy_f_tag = (noisy_f[0:-3]-noisy_f[2:-1])/dx
    label_text = f'lambda = {l}'
    plt.plot(vec[1:-2],noisy_f_tag,label=label_text)
plt.legend(loc='lower right')
plt.show
        
```

As one can see, the noise level in the cental derivative inceasing significally with lambda

#### c

```python
# A p'' = b

vec = np.linspace(0.5,10,101)
dx = vec[2:-1] - vec[0:-3]
f = func(vec)  

intervals = len(vec)-1 # number of interveal

Delta_x = (vec[-1] - vec[0])/intervals # assuming eqaul intervals
M_size = intervals - 1 # Matrix size

# Main diagonal: [4*dx, 4*dx, ...]
main_diag = 4 * Delta_x * np.ones(M_size)

# Off-diagonals (sub and super): [dx, dx, ...]
off_diag = Delta_x * np.ones(M_size - 1)

# Build the tridiagonal matrix A
A = np.diag(main_diag, k=0) + \
    np.diag(off_diag, k=1) + \
    np.diag(off_diag, k=-1)

#print("Coefficient Matrix A:")
#print(A)

# Calculate b vector
b_inner = f[:-2] - 2 * f[1:-1] + f[2:] 
b = (6 / Delta_x) * b_inner

# find p''
p_tag_tag = np.linalg.solve(A, b)
#print(p_tag_tag.shape)
alpha = list()
beta = list()
gamma = beta
etha = gamma
for i in range(len(p_tag_tag)-1):
    alpha.append(p_tag_tag[i+1]/(6*Delta_x))
    beta.append(-p_tag_tag[i]/(6*Delta_x))
    gamma.append((-p_tag_tag[i+1]*Delta_x**2+6*f[i+1])/(6*Delta_x))
    etha.append( (p_tag_tag[i]*Delta_x**2-6*f[i])/(6*Delta_x) )

spline_vec = (np.linspace(0.5,10,1001))
s = 0
P = list()
for x in spline_vec:
    if x>=vec[-3]:
        break
    #print (s)
    P_v = list()
    if x<vec[s+1]: 
        a = alpha[s] * (x-vec[s])**3
        b = beta[s]  * (x-vec[s+1])**3
        g = gamma[s] * (x-vec[s])
        e = etha[s]  * (x-vec[s+1])
        P_v.append(a+b+g+e)
    else:
        s+=1

    P.append(np.array(P_v))

P_arr = np.concatenate(P)
print(P_arr) 

plt.plot(P_arr)
plt.show()



```
